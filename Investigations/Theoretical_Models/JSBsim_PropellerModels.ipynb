```python
# Importing all types of propellers that are available
ae185x12 = prop.aeronaut185x12()
ae20x8 = prop.aeronaut20x8()
ae11x7 = prop.aeronaut11x7_estimatedBEN()

proplist = [ae185x12, ae20x8, ae11x7]
```

```python
# Generating range of values for propeller #1
J = np.linspace(0, 1, 50)
plt.figure()
for i, propeller in enumerate(proplist):
    print("Propeller: "+str(i))
    CT = propeller.thrust_coeff(J)
    CP = ae185x12.power_coeff(J)
    datadict = {'J': J, 'CT': CT}
    datadict = pd.DataFrame(datadict)
    print(datadict.to_string(index=False))
    datadict = {'J': J, 'CP': CP}
    datadict = pd.DataFrame(datadict)
    print(datadict.to_string(index=False))
plt.legend()
plt.show()
```
