```python
from os import sys
import os
sys.path.append('../')

%matplotlib widget

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from aircraft import airplane
import propellers as prop

plt.style.use('../basic_plotter.mplstyle')

# Simple aerodynamic Model of Vehicle
CD0 = 0.018         # Zero lift drag coefficient
k   = 0.03          # Induced drag coefficient (1/(pi*AR*e))

# Static conditions
rho0 = 1.225        # Standard seal level air density for now

createv = airplane(13.0, 0.395, 6.28)

# Linus has wingspan of 2.54 meters and 0.367 m^2 area
linus_chord = 0.367/2.54
linus = airplane(1.82, linus_chord, 2.54)

linus_prop = prop.aeronaut11x7_estimatedBEN()
createv_prop = prop.aeronaut185x12()

V_inf = np.linspace(8, 25.7, 20)      # Airspeed in meters per second to check operating point

q = 0.5 * rho0 * V_inf**2
X_T_createv = (CD0 * q * createv.area) + (k * (q*createv.area)**-1 * createv.weight**2)
X_T_linus = (CD0 * q * linus.area) + (k * (q*linus.area)**-1 * linus.weight**2)

# Getting Estimates for propeller RPM
createv_rpm = np.zeros(np.shape(V_inf))
linus_rpm = np.zeros(np.shape(V_inf))

for i, v in enumerate(V_inf):
    createv_rpm[i] = createv_prop.getRPM(X_T_createv[i], rho0, v)
    linus_rpm[i] = linus_prop.getRPM(X_T_linus[i], rho0, v, n_guess=70)

plt.figure()
plt.plot(V_inf, X_T_createv, label='CREATeV Specifications')
plt.plot(V_inf, X_T_linus, label='Linus Specifications')
plt.legend()
plt.xlabel("$V_\inf$")
plt.ylabel("$X_T$ (N)")
plt.show()

plt.figure()
plt.plot(V_inf, createv_rpm, label='CREATeV Specifications')
plt.plot(V_inf, linus_rpm, label='Linus Specifications')
plt.legend()
plt.xlabel("$V_\inf$ (m/s)")
plt.ylabel("Revs per Second")
plt.show()
```
