```python
import numpy as np
from matplotlib import pyplot as plt

# Getting packages #
%matplotlib widget
```

```python
T0 = 6 # Newton
Tf = 15 # Newton
dt = 0.1 # time interval in seconds
Cd0 = 0.018
k = 0.03
rho = 1.225
S = 2.481
m = 12.6
W = m * 9.807
V_i = 8 # m/s

t = np.arange(1001)
v = np.zeros(np.shape(t))

for i in t:
    V_inew = (dt/m) * (Tf - (0.5 * rho * Cd0 * S * V_i**2) - (2 * k * W**2 * (rho * S * V_i**2)**-1)) + V_i
    v[i] = V_inew
    V_i = V_inew
```

```python
plt.figure()
plt.plot(t*dt, v)
plt.grid()
plt.show()
```
