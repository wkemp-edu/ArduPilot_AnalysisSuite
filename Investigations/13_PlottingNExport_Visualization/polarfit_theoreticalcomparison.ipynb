```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_excel("../../DragBuilds/CREATeV_DragBuild_FreeWake.xlsx")

Cd0 = 0.012
e0 = 0.9

CLdmin = 0.1
AR = 15

CL = np.linspace(-0.2, 1.5)

CD_t1 = Cd0 + (CL**2 * (np.pi * AR * e0)**-1)
CD_t2 = Cd0 + ((CL - CLdmin)**2 * (np.pi * AR * e0)**-1)

plt.figure(dpi=100, figsize=(7,7))
ax = plt.gca()
plt.plot(CD_t1, CL, linestyle='solid', color='black', marker='')
plt.plot(CD_t2, CL, linestyle='solid', color='black', marker='')
plt.scatter(data.CD, data.CL)

# Hide X and Y axes label marks
# ax.xaxis.set_tick_params(labelbottom=False)
# ax.yaxis.set_tick_params(labelleft=False)

ax.set_xlabel("$C_D$")

# Hide X and Y axes tick marks
# ax.set_xticks([])
# ax.set_yticks([])
plt.xlabel("$C_D$")
plt.show()
```
